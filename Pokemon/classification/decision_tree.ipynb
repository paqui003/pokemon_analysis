```python
import pandas as pd 

copy_df = pd.read_csv("../csv/pokemon.csv")
```

```python
copy_df[copy_df["genericLabel"] == "Legendary"]
```

```python
copy_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder

feature_cols = ["Hp", "Atk", "Def", "Satk", "Sdef", "Spd", "Height", "Weight", "Type1", "Type2", "Growthr", "Genderless"]


types_encoded = pd.get_dummies(copy_df[feature_cols], columns = ["Type1", "Type2"])

le_growthr = LabelEncoder() #FLUCTUATING SEEMS TO BE MISPLACED: Careful for distance based machine learning

types_encoded["Growthr"] = le_growthr.fit_transform(types_encoded["Growthr"])
```

```python
le_growthr.classes_
```

```python
types_encoded.columns
```

```python
types_encoded.sample(10)
```

```python
X = types_encoded.copy()
y = copy_df["genericLabel"]


class_le = LabelEncoder()
classes_enc = class_le.fit_transform(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)

dtc = DecisionTreeClassifier()
```

```python
print(f'Legendary frequency in the training set: {sum(y_train == "Legendary")/len(y_train)} \n\
Legendary frequency in the test set: {sum(y_test == "Legendary")/len(y_test)}')
```

```python
param_grid = [{'max_depth': [5, 10, 20], 'max_features': [0.1, 0.25, 0.5, 1.0]}]

dtc_gs = GridSearchCV(dtc, param_grid=param_grid, cv=3, verbose = 3, n_jobs = -1, scoring = "accuracy", return_train_score = True)
```

```python
dtc_gs.fit(X_train, y_train)
```

```python
from sklearn import tree

tree.export_graphviz(dtc_gs.best_estimator_,
                     out_file="../figures/tree.dot",
                     feature_names = X.columns, 
                     class_names=class_le.classes_,
                     filled = True)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import auc

accuracy_score(dtc_gs.best_estimator_.predict(X_test), y_test)
```
